Dataset: https://www.kaggle.com/datasets/thedevastator/airbnb-prices-in-european-cities

### Keterangan Data:

realSum: The total price of the Airbnb listing. (Numeric)

room_type: The type of room being offered (e.g. private, shared, etc.). (Categorical)

room_shared: Whether the room is shared or not. (Boolean)

room_private: Whether the room is private or not. (Boolean)

person_capacity: The maximum number of people that can stay in the room. (Numeric)

host_is_superhost: Whether the host is a superhost or not. (Boolean)

multi: Whether the listing is for multiple rooms or not. (Boolean)

biz: Whether the listing is for business purposes or not. (Boolean)

cleanliness_rating: The cleanliness rating of the listing. (Numeric)

guest_satisfaction_overall: The overall guest satisfaction rating of the listing. (Numeric)

bedrooms: The number of bedrooms in the listing. (Numeric)

dist: The distance from the city centre. (Numeric)

metro_dist	The distance from the nearest metro station. (Numeric)

lng: The longitude of the listing. (Numeric)

lat: The latitude of the listing. (Numeric)

### Gabung Dataset

```python
amwd = pd.read_csv('Data/amsterdam_weekdays.csv', index_col=0)
amwd['city'] = 'Amsterdam'
amwe = pd.read_csv('Data/amsterdam_weekends.csv', index_col=0)
amwe['city'] = 'Amsterdam'

atwd = pd.read_csv('Data/athens_weekdays.csv', index_col=0)
atwd['city'] = 'Athens'
atwe = pd.read_csv('Data/athens_weekends.csv', index_col=0)
atwe['city'] = 'Athens'

barwd = pd.read_csv('Data/barcelona_weekdays.csv', index_col=0)
barwd['city'] = 'Barcelona'
barwe = pd.read_csv('Data/barcelona_weekends.csv', index_col=0)
barwe['city'] = 'Barcelona'

berwd = pd.read_csv('Data/berlin_weekdays.csv', index_col=0)
berwd['city'] = 'Berlin'
berwe = pd.read_csv('Data/berlin_weekends.csv', index_col=0)
berwe['city'] = 'Berlin'

budwd = pd.read_csv('Data/budapest_weekdays.csv', index_col=0)
budwd['city'] = 'Budapest'
budwe = pd.read_csv('Data/budapest_weekends.csv', index_col=0)
budwe['city'] = 'Budapest'

liswd = pd.read_csv('Data/lisbon_weekdays.csv', index_col=0)
liswd['city'] = 'Lisbon'
liswe = pd.read_csv('Data/lisbon_weekends.csv', index_col=0)
liswe['city'] = 'Lisbon'

lonwd = pd.read_csv('Data/london_weekdays.csv', index_col=0)
lonwd['city'] = 'London'
lonwe = pd.read_csv('Data/london_weekends.csv', index_col=0)
lonwe['city'] = 'London'

parwd = pd.read_csv('Data/paris_weekdays.csv', index_col=0)
parwd['city'] = 'Paris'
parwe = pd.read_csv('Data/paris_weekends.csv', index_col=0)
parwe['city'] = 'Paris'

rowd = pd.read_csv('Data/rome_weekdays.csv', index_col=0)
rowd['city'] = 'Rome'
rowe = pd.read_csv('Data/rome_weekends.csv', index_col=0)
rowe['city'] = 'Rome'

viewd = pd.read_csv('Data/vienna_weekdays.csv', index_col=0)
viewd['city'] = 'Vienna'
viewe = pd.read_csv('Data/vienna_weekends.csv', index_col=0)
viewe['city'] = 'Vienna'
```

```python
data_wd = pd.concat([amwd, atwd, barwd, berwd, budwd, liswd, lonwd, parwd, rowd, viewd], ignore_index=True, axis=0)
data_we = pd.concat([amwe, atwe, barwe, berwe, budwe, liswe, lonwe, parwe, rowe, viewe], ignore_index=True, axis=0)
```

```python
data_wd.head()
```

```python
data_wd = data_wd.drop(labels = ["attr_index","attr_index_norm","rest_index","rest_index_norm","lng","lat"], axis = 1)
data_wd["weekend"] = False
data_we = data_we.drop(labels = ["attr_index","attr_index_norm","rest_index","rest_index_norm","lng","lat"], axis = 1)
data_we["weekend"] = True
```

```python
gabungan = pd.concat([data_wd,data_we], ignore_index=True, axis = 0)
```

```python
gabungan.to_csv('datafull.csv')
```
