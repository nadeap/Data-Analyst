```python
data = pd.read_csv('/content/drive/My Drive/dataset/dm/datafull.csv', delimiter=',',decimal='.')
data.head()
```

```python
data.info()
```

```python
data = data.drop(columns='Unnamed: 0')
```

```python
data['CustomerID'] = data['CustomerID'].astype(str)
data['GasStationID'] = data['GasStationID'].astype(str)
data['ProductID'] = data['ProductID'].astype(str)
```

```python
top = data.sort_values(by=['Amount'], ascending=False)
top.head()
```

```python
custtop = top.groupby('CustomerID')['Amount'].sum()
custtop
```

```python
custtop.shape
```

```python
custtop = pd.DataFrame(custtop, columns = ['Amount'])
```

```python
custtop
```

```python
custtop = custtop.sort_values(['Amount'], ascending=False)
custtop.head()
```
